```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
Fuel = pd.read_csv('FuelConsumptionCo2.csv' , sep = ',')
Fuel.head(10)
```

```python
Fuel.describe()
```

```python
num_fuel = Fuel[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]

num_fuel.hist()
plt.show()
```

```python
def scatter_graph(array):
    plt.scatter( array , num_fuel.CO2EMISSIONS )
    plt.xlabel(array.columns[0])
    plt.ylabel('CO2Emissions')
    plt.show()

for column in num_fuel.columns:
    scatter_graph(num_fuel[[column]])
```

```python
isinstance(Fuel['CO2EMISSIONS'], pd.Series)
```

With that we can notice that some graphs have a linear relation with the variable CO2EMISSIONS.
